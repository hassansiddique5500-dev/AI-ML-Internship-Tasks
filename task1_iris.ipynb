```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('iris')

# Print shape
print("Shape:", df.shape)

# Print columns
print("Columns:", df.columns)

# Show first rows
print(df.head())

# Info
print(df.info())

# Describe
print(df.describe())

# Scatter Plot
plt.figure()
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Width")
plt.show()

# Histogram
df.hist()
plt.show()

# Box Plot
plt.figure()
sns.boxplot(data=df)
plt.title("Boxplot of Features")
plt.show()
```
